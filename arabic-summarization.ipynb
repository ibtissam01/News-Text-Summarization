```python
from transformers import BertTokenizer, EncoderDecoderModel
tokenizer = BertTokenizer.from_pretrained("Goud/AraBERT-summarization-goud")
model = EncoderDecoderModel.from_pretrained("Goud/AraBERT-summarization-goud")

def generate_summary(article):

    input_ids = tokenizer(article, return_tensors="pt", truncation=True, padding=True).input_ids
    generated = model.generate(input_ids)[0]
    output = tokenizer.decode(generated, skip_special_tokens=True)
    
    return output

```

```python
# Set the device to GPU if available, otherwise use CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the tokenizer and model
tokenizer = BertTokenizer.from_pretrained("Goud/AraBERT-summarization-goud")
model = EncoderDecoderModel.from_pretrained("Goud/AraBERT-summarization-goud")
model = model.to(device)

# Define the function to generate summaries
def generate_summary(article):
    input_ids = tokenizer(article, return_tensors="pt", truncation=True, padding=True).input_ids.to(device)
    generated = model.generate(input_ids)[0]
    output = tokenizer.decode(generated, skip_special_tokens=True)
    return output

# Apply the generate_summary function to the "texte" column
df["resume_genere"] = df["texte"].apply(generate_summary)

# Save the DataFrame with ROUGE scores to a CSV file
df.to_csv("Arabert_resume_scores.csv", index=False)
```

```python
from rouge import Rouge
# Define the function to evaluate summaries using ROUGE
def evaluate_summary(generated_summary, reference_summary):
    rouge = Rouge()
    scores = rouge.get_scores(generated_summary, reference_summary)
    return scores

# Iterate over each row in the DataFrame and calculate ROUGE scores
evaluation_scores = []
for index, row in df.iterrows():
    generated_summary = row["resume_genere"]
    reference_summary = row["summary"]
    scores = evaluate_summary(generated_summary, reference_summary)
    evaluation_scores.append(scores)

# Add the ROUGE scores to the DataFrame
df["rouge_scores"] = evaluation_scores
# Save the DataFrame with ROUGE scores to a CSV file
df.to_csv("Arabert_resume_scores.csv", index=False)
```

```python
# Calculer la moyenne des scores ROUGE pour chaque métrique
rouge_1_scores = df['rouge_scores'].apply(lambda x: x[0]['rouge-1']['f']).mean()
rouge_2_scores = df['rouge_scores'].apply(lambda x: x[0]['rouge-2']['f']).mean()
rouge_l_scores = df['rouge_scores'].apply(lambda x: x[0]['rouge-l']['f']).mean()

# Afficher les résultats
print("ROUGE-1 score moyen :", rouge_1_scores)
print("ROUGE-2 score moyen :", rouge_2_scores)
print("ROUGE-L score moyen :", rouge_l_scores)
```

En résumé, les scores ROUGE moyens indiquent une correspondance relativement faible entre les résumés générés et les résumés de référence en termes de mots individuels et de paires de mots consécutifs.

```python
# Define the function to evaluate summaries using BLEU
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
# Define the function to evaluate summaries using BLEU
def evaluate_summary(generated_summary, reference_summary):
    reference_tokens = reference_summary.split()
    generated_tokens = generated_summary.split()
    smoothing_function = SmoothingFunction().method1  # Define smoothing function
    bleu_score = sentence_bleu([reference_tokens], generated_tokens, smoothing_function=smoothing_function)
    return bleu_score

# Iterate over each row in the DataFrame and calculate BLEU scores
evaluation_scores = []
for index, row in df.iterrows():
    generated_summary = row["resume_genere"]
    reference_summary = row["summary"]
    score = evaluate_summary(generated_summary, reference_summary)
    evaluation_scores.append(score)

# Add the BLEU scores to the DataFrame
df["bleu_scores"] = evaluation_scores

# Save the DataFrame with BLEU scores to a CSV file
df.to_csv("Arabert_resume_scores.csv", index=False)
```

```python
# Load the DataFrame with BLEU scores
df = pd.read_csv("Arabert_resume_scores.csv")

# Calculate the average BLEU score
average_bleu_score = df["bleu_scores"].mean()

# Print the average BLEU score
print("Average BLEU score:", average_bleu_score)
```

The average BLEU score of approximately 0.0052 suggests that the generated summaries have a low level of similarity to the reference summaries. The BLEU score ranges from 0 to 1, with a higher score indicating a better match between the generated and reference summaries.
